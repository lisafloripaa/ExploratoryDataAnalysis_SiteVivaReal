# Análise Exploratória de Dados  (Exploratory Data Analysis)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
df = pd.read_csv('full_results_clean.csv')
```

### Identificando e removendo outliers (Identifying and removing outliers)

```python
df.describe()
```

```python
df[['garages']].boxplot()
```

```python
df = df[df.garages <= 10]
```

```python
df.describe()
```

```python
df[['price']].boxplot()
```

```python
df = df[df.price <= 15000]
```

```python
df.describe()
```

```python
df[['condo']].boxplot()
```

```python
df = df[df.condo <= 10000]
```

```python
df.describe()
```

```python
df[['area']].boxplot()
```

```python
df = df[df.area <= 300]
```

```python
df.describe()
```

### Visualizando Distribuicões e Correlações (Viewing Distributions and Correlations)

```python
plt.figure(figsize=(10,6))
plt.scatter(df['area'],df['price'], color = 'k', alpha=0.2)
plt.ylim([0,20000])
plt.xlabel('Área',fontsize=18)
plt.ylabel('Preço do aluguel',fontsize=18)
```

```python
sns.pairplot(df, kind='reg')
```

```python
df.corr()
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
```

```python
sns.boxplot(x='zona', y='price', data=df)
```

```python
sns.pairplot(df, hue='zona')
```

```python
ax = sns.FacetGrid(df, col='zona', height=6.5)
ax.map(plt.scatter, 'area','price',alpha=0.5)
```

```python
df1 = df.loc[df['zona'] == 'Central']
df2 = df.loc[df['zona'] == 'Continental']
df3 = df.loc[df['zona'] == 'Norte']
df4 = df.loc[df['zona'] == 'Sul']
df5 = df.loc[df['zona'] == 'Leste']



trace1 = go.Scatter(x = df1.area,
                   y = df1.price,
                   mode='markers', 
                   name='Central',
                   marker = dict(color = 'rgba(255,128,255,0.8)'),
                    text=df1.zona)

trace2 = go.Scatter(x = df2.area,
                   y = df2.price,
                   mode='markers', 
                   name='Continetal',
                   marker = dict(color = 'rgba(181,149,37,0.8)'),
                    text=df2.zona)

trace3 = go.Scatter(x = df3.area,
                   y = df3.price,
                   mode='markers', 
                   name='Norte',
                   marker = dict(color = 'rgba(11,156,49,0.8)'),
                    text=df3.zona)

trace4 = go.Scatter(x = df4.area,
                   y = df4.price,
                   mode='markers', 
                   name='Sul',
                   marker = dict(color = 'rgba(40,115,138,0.8)'),
                    text=df4.zona)

trace5 = go.Scatter(x = df5.area,
                   y = df5.price,
                   mode='markers', 
                   name='Leste',
                   marker = dict(color = 'rgba(55,70,73,0.8)'),
                    text=df5.zona)

data = [trace1, trace2, trace3, trace4, trace5]
layout = dict(title='Área e Preço do Aluguel de Apartamentos em Florianópolis - SC',
             xaxis=dict(title='Área'),
             yaxis=dict(title='Preço'))

fig = dict(data = data, layout=layout)
iplot(fig)
```
